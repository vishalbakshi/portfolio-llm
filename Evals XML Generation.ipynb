```python
import pandas as pd
```

```python
df = pd.read_csv("evals.csv")
df
```

```python
item = df.iloc[0]
item
```

```python
itemstr = """<item>
<model>{model}</model>
<prompt-type>{prompt_type}<prompt-type>
<user>{user}</user>
<prompt>{prompt}</prompt>
<response>{response}</response>
<gold-answer>{gold_answer}</gold-answer>
<completeness>{completeness}</completeness>
<accuracy>{accuracy}</accuracy>
<conciseness>{conciseness}</conciseness>
<hallucinations>{hallucinations}</hallucinations>
<format-adherence>{format_adherence}</format-adherence>
</item>"""
```

```python
print(itemstr.format(
    model=item["Model"],
    prompt_type=item["Prompt Type"],
    user=item["User"],
    prompt=item["Prompt"],
    response=item["Response"],
    gold_answer=item["Gold Answer"],
    completeness=item["Completeness"],
    accuracy=item["Accuracy"],
    conciseness=item["Conciseness"],
    hallucinations=item["Hallucinations"],
    format_adherence=item["Format Adherence"]
))
```

```python
jsonstr = "<dataset>\n"

for index, item in df.iterrows():
    jsonstr += itemstr.format(
        model=item["Model"],
        prompt_type=item["Prompt Type"],
        user=item["User"],
        prompt=item["Prompt"],
        response=item["Response"],
        gold_answer=item["Gold Answer"],
        completeness=item["Completeness"],
        accuracy=item["Accuracy"],
        conciseness=item["Conciseness"],
        hallucinations=item["Hallucinations"],
        format_adherence=item["Format Adherence"]
    )

jsonstr += "\n</dataset>"
```

```python
print(jsonstr)
```

```python
with open("evals.txt", "w") as f:
    f.write(jsonstr)
```
